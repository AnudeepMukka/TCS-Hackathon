```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
Survey_data = pd.read_csv("survey.csv")
```

```python
def Clean_Gender(row):
    male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "cis male"]
    female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]
    if row['Gender'].lower() in male_str:
        return 'Male'
    elif row['Gender'].lower() in female_str:
        return 'Female'
    else:
        return 'Transgender'
```

```python
Survey_data['clean_gender'] = Survey_data.apply(Clean_Gender,axis = 1)
```

```python
def clean_Age(df, Median_age):
    Age= df["Age"]
    print(Age)
    #Converting negative numbers
    Age = abs(Age) 
    
    # Fill with median() values < 18 and > 120
    if (Age < 18) | (Age > 120):
        Age = Median_age
    return Age
```

```python
Survey_data["clean_Age"] = Survey_data.apply(clean_Age, axis = 1, Median_age = Survey_data.Age.median())
```

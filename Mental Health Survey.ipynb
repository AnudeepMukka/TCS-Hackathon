```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
Survey_data = pd.read_csv("survey.csv")
```

```python
def Clean_Gender(row):
    male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "cis male"]
    female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]
    if row['Gender'].lower() in male_str:
        return 'Male'
    elif row['Gender'].lower() in female_str:
        return 'Female'
    else:
        return 'Transgender'
```

```python
Survey_data.columns
```

```python
Survey_df = Survey_data
```

```python
Survey_df['Gender'] = Survey_data.apply(Clean_Gender,axis = 1)
```

```python
def clean_Age(df, Median_age):
    Age= df["Age"]
    print(Age)
    #Converting negative numbers
    Age = abs(Age) 
    
    # Fill with median() values < 18 and > 120
    if (Age < 18) | (Age > 120):
        Age = Median_age
    return Age
```

```python
Survey_df["Age"] = Survey_data.apply(clean_Age, axis = 1, Median_age = Survey_data.Age.median())
```

```python
Survey_df = Survey_df.drop(["Timestamp"], axis = 1)
Survey_df = Survey_df.drop(["comments"], axis = 1)
Survey_df = Survey_df.drop(["Country"], axis = 1)
Survey_df = Survey_df.drop(["state"], axis = 1)
```

```python
labelDict = {}
for feature in features:
    le = LabelEncoder()
    le.fit(Survey_df[feature])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    Survey_df[feature] = le.transform(Survey_df[feature])
    # Get labels
    labelKey = 'label_' + feature
    labelValue = [*le_name_mapping]
    labelDict[labelKey] =labelValue
    
for key, value in labelDict.items():     
    print(key, value)
```
